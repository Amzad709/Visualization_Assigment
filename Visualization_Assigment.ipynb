```python
1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic
dataset.



##Code



np.random.seed(42)  

# Generate x values (e.g., from 0 to 100)
x = np.linspace(0, 100, 100)


slope = 3
intercept = 10
noise = np.random.normal(0, 10, size=x.shape)  
y = slope * x + intercept + noise


plt.figure(figsize=(10, 6)) 
plt.scatter(x, y, color='blue', marker='o', edgecolor='black', alpha=0.7)


plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('x')
plt.ylabel('y')


plt.grid(True)

# Display the plot
plt.show()


```

```python
3. Create a dataset representing categories and their corresponding values. Compare different categories
based on numerical values.

##Code



np.random.seed(42)  


categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = {cat: np.random.normal(loc=10, scale=5, size=100) for cat in categories}

data = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([values[cat] for cat in categories])
})


plt.figure(figsize=(12, 6))
mean_values = data.groupby('Category')['Value'].mean()
mean_values.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Mean Value by Category')
plt.xlabel('Category')
plt.ylabel('Mean Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Box Plot: Distribution of Values for Each Category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Value', data=data, palette='Set2')
plt.title('Distribution of Values by Category')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

##Code


np.random.seed(0)

# Number of samples
n_samples = 100

# synthetic data with correlation
# Feature1 is random
feature1 = np.random.normal(loc=0, scale=1, size=n_samples)


feature2 = 0.7 * feature1 + np.random.normal(loc=0, scale=1, size=n_samples)

feature3 = 0.5 * feature1 + 0.3 * feature2 + np.random.normal(loc=0, scale=1, size=n_samples)


df = pd.DataFrame({
    'Feature1': feature1,
    'Feature2': feature2,
    'Feature3': feature3
})


correlation_matrix = df.corr()

# heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()

```

```python
1.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.


##Code#####


np.random.seed(42)

# Generate random data
n = 100  # Number of points
x = np.random.uniform(0, 100, n)
y = np.random.uniform(0, 100, n)
sizes = np.random.uniform(5, 50, n)  # Marker sizes
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], n)  # Marker colors

# Create a DataFrame
df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Prepare the data for Bokeh
source = ColumnDataSource(df)

# Display the plot inline in a Jupyter notebook
output_notebook()

# Create a scatter plot
p = figure(title="Scatter Plot with Different Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Add a scatter renderer
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, legend_field='colors', fill_alpha=0.6)

# Show the plot
show(p)
```
